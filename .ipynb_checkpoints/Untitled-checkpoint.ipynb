```python
import numpy as np
import pandas as pd
from scipy import optimize
```

```python
def create_data(n_obs):
    x_1 = np.random.uniform(0, 10, n_obs)
    x_2 = np.random.uniform(-5, 5, n_obs)
    e = np.random.normal(0, 1, n_obs)
    y = 2 + 3*x_1 + 5*x_2 + e
    dat ={'x1': x_1, 'x2': x_2, 'y': y}
    df = pd.DataFrame(dat)
    return df
    
```

```python
df = create_data(100)
```
